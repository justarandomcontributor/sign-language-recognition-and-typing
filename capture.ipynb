```python
import cv2
import numpy as np
import os 

def empty(a):
    pass

# code to stack windows horizontally or vertically 
# taken from https://youtu.be/WQeoO7MI0Bs also can refer to the link in the video description

def stackImages(scale,imgArray):
    rows = len(imgArray)
    cols = len(imgArray[0])
    rowsAvailable = isinstance(imgArray[0], list)
    width = imgArray[0][0].shape[1]
    height = imgArray[0][0].shape[0]
    if rowsAvailable:
        for x in range ( 0, rows):
            for y in range(0, cols):
                if imgArray[x][y].shape[:2] == imgArray[0][0].shape [:2]:
                    imgArray[x][y] = cv2.resize(imgArray[x][y], (0, 0), None, scale, scale)
                else:
                    imgArray[x][y] = cv2.resize(imgArray[x][y], (imgArray[0][0].shape[1], imgArray[0][0].shape[0]), None, scale, scale)
                if len(imgArray[x][y].shape) == 2: imgArray[x][y]= cv2.cvtColor( imgArray[x][y], cv2.COLOR_GRAY2BGR)
        imageBlank = np.zeros((height, width, 3), np.uint8)
        hor = [imageBlank]*rows
        hor_con = [imageBlank]*rows
        for x in range(0, rows):
            hor[x] = np.hstack(imgArray[x])
        ver = np.vstack(hor)
    else:
        for x in range(0, rows):
            if imgArray[x].shape[:2] == imgArray[0].shape[:2]:
                imgArray[x] = cv2.resize(imgArray[x], (0, 0), None, scale, scale)
            else:
                imgArray[x] = cv2.resize(imgArray[x], (imgArray[0].shape[1], imgArray[0].shape[0]), None,scale, scale)
            if len(imgArray[x].shape) == 2: imgArray[x] = cv2.cvtColor(imgArray[x], cv2.COLOR_GRAY2BGR)
        hor= np.hstack(imgArray)
        ver = hor
    return ver


cap = cv2.VideoCapture(1)  #start video capture
count=0
flag=True
while flag:  
    suc, img = cap.read()  # extract the frames one by one from captured video and store them in "img"
#     imgCANNY = cv2.Canny(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY),100,200)  # CANNY filter applied on the GRAY SCALE version of captured image
    img2= img.copy() # copy of original captured image
    
#     print(np.array(imgCANNY).shape, np.array(img).shape)
    cv2.rectangle(img2,(20,0),(260,240),(0,255,0),3)  # draw the focus area on the copy image
    
    sign = img[0:240,20:260]
    
    key = cv2.waitKey(1)   # store the key pressed if not stores -1
    if key == ord('q'):    # if "q" is pressed then capture the focused image area with Canny filter applied
        name = str(count)+".jpg"  # name for the image to be saved
        
#         path= "C:\\Users\\sinha\\Desktop\\dataset\\a"  # the directory to store the captured image
        path2 = "C:\\Users\\sinha\\Desktop\\dataset\\a_original"  # the directory to save original image
        
#         cv2.imwrite(os.path.join(path , name),imgCANNY[220:460,90:550])  # writting to the directory
        cv2.imwrite(os.path.join(path2 , name),sign)  # writting to the second directory
        count=count+1
        
    elif key == ord('z') or count>=250:  # if "z" is pressed the close the capture proccess
        flag=False
        cap.release()  # closes the video capture
        
    imgStack = stackImages(0.6,[img2,sign])
    cv2.imshow("Stacked Images", imgStack)   # displaying the stacked images
    
cv2.destroyAllWindows()  # destroy/close the windows opened by the program
```
